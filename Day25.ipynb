```python
with open("input25.txt") as f:
    inp = [x[:-1] for x in f.readlines()]
    
data = dict()
for i, line in enumerate(inp):
    for j, sym in enumerate(line):
        if sym in ['>','v']:
            data[(i,j)] = sym
nrows = len(inp)
ncols = len(inp[0])

data
```

# Part 1

```python
def move_east(data):
    new_data = dict()
    for i,j in data:
        if data[(i,j)] == '>':
            if (i, (j+1)%ncols) not in data.keys():
                new_data[(i, (j+1)%ncols)] = '>'
            else: 
                new_data[(i,j)] = '>'
        else:
            new_data[(i,j)] = 'v'
    return new_data

def move_south(data):
    new_data = dict()
    for i,j in data:
        if data[(i,j)] == 'v':
            if ((i+1)%nrows, j) not in data.keys():
                new_data[((i+1)%nrows, j)] = 'v'
            else: 
                new_data[(i,j)] = 'v'
        else:
            new_data[(i,j)] = '>'
    return new_data

def one_step(data):
    data = move_east(data)
    data = move_south(data)
    return data

def solve(data):
    keys = data.keys()
    count = 0
    while True:
        data = one_step(data)
        count +=1
        if data.keys() == keys:
            return count
        keys = data.keys()

solve(data)
```

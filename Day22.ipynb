```python
import re
from time import time

with open("input22.txt") as f:
    inp = [x[:-1] for x in f.readlines()]
data = [re.findall(r'(on|off).*x=(.*).*\.\.(.*),y=(.*)\.\.(.*),z=(.*)\.\.(.*)', x)[0] for x in inp]
data = [(x,int(y), int(z), int(t), int(u), int(w), int(o)) for x,y,z,t,u,w,o in data]
data
```

# Part 1

```python
init = [(x,y,z,t,u,v,w) for x,y,z,t,u,v,w in data if -50 <= min(y,z,t,u,v,w) and max(y,z,t,u,v,w) <= 50]


def on_off(line, on):
    blocks = [(i,j,k) for i in range(line[1], line[2]+1) for j in range(line[3], line[4]+1) for k in range(line[5], line[6]+1)]
    if line[0] == 'on':
        for block in blocks:
            if block not in on.keys():
                on[block] = 1
    if line[0] == 'off':
        for block in blocks:
            if block in on.keys():
                del on[block]
    return on

def part1(init):
    on = dict()
    for line in init:
        print(line)
        on = on_off(line, on)
    return on

len(part1(init[:12]).keys())
```

# Part 2

```python
def int_cubes(c1,c2, switch='on'):
    """Gives a disjoint set of rect prisms st its union is c1 U c2"""
    if switch == 'off':
        res = int_cubes(c1, c2, 'on')
        res.remove(c1)
        return res
    if (c2[1] < c1[0] or c2[0] > c1[1] \
           or c2[3] < c1[2] or c2[2] > c1[3]\
           or c2[5] < c1[4] or c2[4] > c1[5]):
        return {c1,c2}
    #Decompose c2 into prisms
    prisms = {c1}
    if c2[0] < c1[0]:
        prisms.add((c2[0], c1[0]-1, c2[2], c2[3], c2[4], c2[5]))
        c2 = (c1[0], c2[1],c2[2],c2[3],c2[4],c2[5])
        return prisms.union(int_cubes(c1, c2))
    if c2[1] > c1[1]:
        prisms.add((c1[1]+1, c2[1], c2[2], c2[3], c2[4], c2[5]))
        c2 = (c2[0], c1[1],c2[2],c2[3],c2[4],c2[5])
        return prisms.union(int_cubes(c1, c2))
    if c2[2] < c1[2]:
        prisms.add((c2[0], c2[1], c2[2],c1[2]-1,c2[4],c2[5]))
        c2 = (c2[0],c2[1],c1[2],c2[3], c2[4], c2[5])
        return prisms.union(int_cubes(c1, c2))
    if c2[3] > c1[3]:
        prisms.add((c2[0], c2[1], c1[3]+1, c2[3], c2[4], c2[5]))
        c2 = (c2[0], c2[1],c2[2],c1[3],c2[4],c2[5])
        return prisms.union(int_cubes(c1, c2))
    if c2[4] < c1[4]:
        prisms.add((c2[0], c2[1], c2[2],c2[3],c2[4],c1[4]-1))
        c2 = (c2[0],c2[1],c2[2],c2[3], c1[4], c2[5])
        return prisms.union(int_cubes(c1, c2))
    if c2[5] > c1[5]:
        prisms.add((c2[0], c2[1], c2[2], c2[3], c1[5]+1, c2[5]))
        c2 = (c2[0], c2[1],c1[2],c2[3],c2[4],c1[5])
        return prisms.union(int_cubes(c1, c2))
    else:
        return {c1}

def int_cube_prisms(cube, prisms):
    new_prisms = set()
    for p in prisms:
        new_prisms = new_prisms.union(int_cubes(cube[1:], p, cube[0]))
    return new_prisms

def decompose(data):
    prisms = {data[0][1:]}
    for cube in data[1:]:
        prisms = int_cube_prisms(cube, prisms)    
    return prisms

def count_cubes(prisms):
    tot = 0
    for p in prisms:
        tot += (p[1]-p[0]+1)*(p[3]-p[2]+1)*(p[5]-p[4]+1)
    return tot

count_cubes(decompose(data))
```

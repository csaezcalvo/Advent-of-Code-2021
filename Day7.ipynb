# Part 1

```python
import numpy as np
sum([np.abs(x-np.median(inp)) for x in inp])
```

# Part 2

```python
def least_fuel(inp):
    least = 0
    least_sum = sum([np.abs(x)*(np.abs(x)+1)/2 for x in inp])
    for i in range(1, max(inp)+1):
        suma = sum([np.abs(x-i)*(np.abs(x-i)+1)/2 for x in inp])
        if suma < least_sum:
            least_sum=suma
            least=i
    return least, least_sum

least_fuel(inp)
```

# Part 1

```python
class Fish():
    def __init__(self, new=True, days=8):
        self.days = days
        self.new = new
    
    def update(self):
        self.days -= 1
        if self.days < 0:
            self.days = 6

def create_in_fish(in_days):
    fishes = []
    for x in in_days:
        fishes.append(Fish(False, x))
    return fishes

def update_list(fishes):
    new = 0
    for fish in fishes:
        if fish.days == 0:
            new += 1
        fish.update()
    return fishes + [Fish() for i in range(new)]

def simulate(fishes, n):
    for i in range(n):
        fishes = update_list(fishes)
    return fishes

#in_days = [3,4,3,1,2]
fishes = create_in_fish(in_days)
fishes = simulate(fishes, 80)
len(fishes)
```

# Part 2

Suppose only one fish start in initial state with 8 days remaining. Then, in n days there will be:
$$N(n) = 1 + \sum_{i=0}^\infty N(9-7i)$$
fishes (where N(n)=0 for n<0).
Since different fishes are independent, an initial fish with $a$ days in initial state (and $d$ days left) is equivalent to a fish with $8$ days in initial state and $d+8-i$ days left. So we simulate each of our fishes in initial state and sum all fishes produced.

```python
def recursion_lista(n, lista):
    if n<0:
        return 0
    return 1 + sum([lista[n-9-7*i] for i in range((n-9)//7+1) if i>= 0])

def recursion(n):
    lista = []
    for i in range(n+1):
        lista.append(recursion_lista(i, lista))
    return lista[-1]

recursion(256)
```

```python
def recursion(n):
    values = []
    for i in range(n+1):
        values.append(1 + sum([values[i-9-7*j] for j in range((n-9)//7+1) if i-9-7*j in range(0,len(values))]))
    return values
    

def sol(in_state, days):
    rem_days = [days+8-i for i in in_state]
    values = recursion(max(rem_days))
    return sum([values[i] for i in rem_days])

#Part 1
sol(in_state, 80)
#Part 2
sol(in_state, 256)
```

# Recursive solution

```python
f = open("input6.txt")
inp = [x[:-1] for x in f.readlines()]
f.close()
in_state = [int(x) for x in inp[0].split(',')]

def recursion(n):
    values = []
    for i in range(n+1):
        values.append(1 + sum([values[i-9-7*j] for j in range((n-9)//7+1) if i-9-7*j in range(len(values))]))
    return values

def sol(in_state, days):
    rem_days = [days+8-i for i in in_state]
    values = recursion(max(rem_days))
    return sum([values[i] for i in rem_days])

#Part 1
sol(in_state, 80)
#Part 2
sol(in_state, 256)
```

# Another way: dictionaries

```python
f = open("input6.txt")
inp = [x[:-1] for x in f.readlines()]
f.close()
in_state = [int(x) for x in inp[0].split(',')]

days_dict = {i:sum([x == i for x in in_state]) for i in range(9)}

def sol(d, days):
    for _ in range(days):
        d = {i:d[(i+1)%9] if i != 6 else d[0]+d[7] for i in range(9)}
    return sum(d.values())

#Part 1
sol(days_dict, 80)
#Part 1
sol(days_dict, 256)
```

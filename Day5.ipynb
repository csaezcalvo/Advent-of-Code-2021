- Mejorar el algoritmo para obtener puntos (sobre todo diagonales)

```python
f = open("input5.txt")
inp = [x[:-1] for x in f.readlines()]
f.close()
inp
```

# Part 1

```python
import re
lines = []
for x in inp:
    lines.append([int(y) for y in re.findall(r'(\d+),(\d+) -> (\d+),(\d+)',x)[0]])
lines
```

```python
def points_in_line(line):
    return [(x,y) for x in range(min(line[0], line[2]), max(line[0],line[2])+1)\
            for y in range(min(line[1], line[3]), max(line[1],line[3])+1)]

hor_vert_lines = [points_in_line(line) for line in lines if line[0]==line[2] or line[1]==line[3]]
hor_vert_lines[0]
```

```python
def part1(lines):
    nlines = dict()
    for line in lines:
        for pos in line:
            if pos not in nlines.keys():
                nlines[pos] = 1
            else:
                nlines[pos] += 1
    return sum([x >= 2 for x in nlines.values()])

part1(hor_vert_lines)
```

# Part 2

```python
def points_in_line2(line):
    minx = min(line[0], line[2])
    maxx = max(line[0],line[2])
    miny = min(line[1], line[3])
    maxy = max(line[1],line[3])
    if minx==maxx or miny==maxy:
        return points_in_line(line)
    else:
        if line[0]<line[2] and line[1]<line[3]:
            return [(line[0]+i, line[1]+i) for i in range(line[2]-line[0]+1)]
        elif line[0]>line[2] and line[1]>line[3]:
            return [(line[0]-i, line[1]-i) for i in range(line[0]-line[2]+1)]
        elif line[0]<line[2] and line[1]>line[3]:
            return [(line[0]+i, line[1]-i) for i in range(line[2]-line[0]+1)]
        else:
            return [(line[0]-i, line[1]+i) for i in range(line[0]-line[2]+1)]
        
points_in_line2([7,9,9,7])
part1([points_in_line2(line) for line in lines])
```

```python
with open("input10.txt") as f:
    inp = [x[:-1] for x in f.readlines()]
inp
```

# Part 1

```python
open_close = {'[':']', '(':')', '{':'}', '<':'>'}
score = {')':3, ']':57, '}':1197, '>':25137}

def end_line(line):
    symbols = [x for x in line]
    heap = []
    for x in symbols:
        if x in open_close.keys():
            heap.append(x)
        elif x == open_close[heap[-1]]:
            heap.pop()
        else:
            return x
    return heap

def error_score(line):
    x = end_line(line)
    if type(x)==str:
        return score[x]
    return 0

sum([error_score(line) for line in inp])
```

# Part 2

```python
inc_score = {')':1, ']':2, '}':3, '>':4}

def end_seq(line):
    heap = end_line(line)
    if type(heap) != str:
        return [open_close[x] for x in heap][::-1]
    return 0

def score(line):
    heap = end_seq(line)
    if heap == 0:
        return 0
    scores = [inc_score[x] for x in heap]
    score = 0
    for i in scores:
        score *= 5
        score += i
    return score

def solve(inp):
    x = [score(line) for line in inp if score(line) != 0]
    return sorted(x)[(len(x)-1)//2]

solve(inp)
```

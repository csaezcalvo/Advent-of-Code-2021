# Part 1

```python
import re

f = open("input2.txt")
inp = [line[:-1] for line in f.readlines()]
f.close()

moves = [re.findall(r'(\w+) (\d+)', x)[0] for x in inp]
moves
```

```python
def final_pos(in_pos, moves):
    for move in moves:
        if move[0] == 'forward':
            in_pos[0] += int(move[1])
        elif move[0] == 'down':
            in_pos[1] += int(move[1])
        else:
            in_pos[1] -= int(move[1])
    return in_pos

res=final_pos([0,0], moves)
res[0]*res[1]
```

# Part 2

```python
def final_pos2(in_pos, moves):
    aim = 0
    for move in moves:
        if move[0] == 'forward':
            in_pos[0] += int(move[1])
            in_pos[1] += int(move[1])*aim
        elif move[0] == 'down':
            aim += int(move[1])
        else:
            aim -= int(move[1])
    return in_pos

res2 = final_pos2([0,0], moves)
res2[0]*res2[1]
```

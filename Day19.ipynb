# Part 1

Scanner 0 is at (0,0,0) and its orientation defines x,y,z axis and positive directions. All other positions will be relative to this one. 
Idea:
Fix a beacon in scanner 0. Then, assume it coincides with a beacon in scanner 1 (for some of the 24 orientations), compute the matrix of positions of scanner 1 relative to scanner 0 (absolute coord) and in the given orientation, and check if there are at least 12 beacons that match. If that is the case, we have identified position and orientation of scanner 1. Otherwise, repeat with next beacon of scanner 1.

```python
def orientations(vector):
    """Returns the positions of the vector for each of the 24 possible orientations"""
    orient = []
    vx, vy, vz = vector[0], vector[1], vector[2]
    return [np.array([vx,vy,vz]), np.array([vy,vz,vx]), np.array([vz,vx,vy]),\
           np.array([-vx,vz,vy]), np.array([vz,vy,-vx]), np.array([vy,-vx,vz]),\
           np.array([vx,-vz,vy]), np.array([-vz,vy,vx]), np.array([vy,vx,-vz]),\
           np.array([vx,vz,-vy]), np.array([vz,-vy,vx]), np.array([-vy,vx,vz]),\
           np.array([-vx,-vy,vz]), np.array([-vy,vz,-vx]), np.array([vz,-vx,-vy]),\
           np.array([vx,-vy,-vz]), np.array([-vy,-vz,vx]), np.array([-vz,vx,-vy]),\
           np.array([-vx,vy,-vz]), np.array([vy,-vz,-vx]), np.array([-vz,-vx,vy]),\
           np.array([-vx,-vz,-vy]), np.array([-vz,-vy,-vx]), np.array([-vy,-vx,-vz])]

def orient_all(scan):
    part = [orientations(x) for x in scan]
    return [[x[i] for x in part] for i in range(24)]
    
def rel_pos(pos1, pos2):
    """Given 2 positions which are the same in diff coord, compute pos of the origin of 2 wrt 1"""
    return pos1 - pos2

def find_pos(scan1, scan2):
    orient_list = orient_all(scan2)
    for i in scan1[:-12]:
        for orient in orient_list:
            for j in orient:
                pos = rel_pos(i,j)
                new_coord = [pos + x for x in orient]
                if sum([all(y==z) for y in scan1 for z in new_coord])>=12:
                    return pos, new_coord
    return False

st=time()
find_pos(data[0], data[1])
end=time()
end-st
```

```python
def solve(data):
    n = len(data)
    pos_list = {0:np.array([0,0,0])}
    coords = {0: data[0]}
    done = []
    while sorted(list(pos_list.keys())) != list(range(n)):
        if [x for x in pos_list.keys() if x not in done]:
            ind1 = [x for x in pos_list.keys() if x not in done][0]
        st = time()
        for ind2 in [x for x in range(n) if x not in pos_list.keys()]:
            print('searching', ind1, ind2)
            res = find_pos(coords[ind1], data[ind2])
            if res:
                print('match!', ind1, ind2)
                pos_list[ind2] = res[0]
                coords[ind2] = res[1]
        end = time()
        print(pos_list.keys())
        print(end-st)
        done.append(ind1)
    return pos_list, coords

st = time()
pos_list, coords = solve(data)
end = time()
end-st

#matches = [(0,6), (0,14), (0,24), (24,2), (24,13), (2,1), (2,22), (13,5), (13,9), (13,18), (1,15), (5,16),\
#          (5,26), (5,28), (9,12), (18,8), (18,11), (15,3), (16,17), (26,7), (26,10), (28,20), (28,23), (28,27),\
#          (8,21), (11,4), (27,19), (19,25)]
```

```python
def manhattan(p1,p2):
    return abs(p1[0]-p2[0]) + abs(p1[1]-p2[1]) + abs(p1[2]-p2[2])

def max_dist(pos_list):
    res = 0
    for i in pos_list.values():
        for j in pos_list.values():
            part = manhattan(i,j)
            if part > res:
                res = part
    return res

max_dist(pos_list)
```

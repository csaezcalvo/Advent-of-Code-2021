# Part 1

```python
# Helper functions
def most_common(l):
    """Returns most common element in l"""
    el_dict = dict()
    for el in l:
        if el not in el_dict.keys():
            el_dict[el] = 1
        else:
            el_dict[el] += 1
    return max(el_dict.items(), key = lambda x: x[1])[0]

def bin_to_dec(x):
    """Converts a string in binary to a decimal number"""
    return sum([int(x[i])*2**(len(x)-i-1) for i in range(len(x))])

bin_to_dec('111')
```

```python
def gamma_rate(inp):
    l = len(inp[0])
    gamma = ''
    for i in range(l):
        gamma = gamma + most_common([x[i] for x in inp])
    return gamma

def epsilon_rate(inp):
    gamma = gamma_rate(inp)
    eps = ''
    for i in range(len(gamma)):
        if gamma[i] == '0':
            eps += '1'
        else:
            eps += '0'
    return eps

bin_to_dec(gamma_rate(inp))*bin_to_dec(epsilon_rate(inp))
```

# Part 2

```python
def count(l):
    """Returns dict with element count in l"""
    el_dict = dict()
    for el in l:
        if el not in el_dict.keys():
            el_dict[el] = 1
        else:
            el_dict[el] += 1
    return el_dict

def bin_to_dec(x):
    """Converts a string in binary to a decimal number"""
    return sum([int(x[i])*2**(len(x)-i-1) for i in range(len(x))])

count([x[1] for x in inp])
```

```python
def oxygen(inp):
    i=0
    while len(inp)>1:
        el_count = count([x[i] for x in inp])
        if el_count['1'] >= el_count['0']:
            inp = [x for x in inp if x[i]=='1']
        else:
            inp = [x for x in inp if x[i]=='0']
        i += 1
    return inp[0]

def co2(inp):
    i=0
    while len(inp)>1:
        el_count = count([x[i] for x in inp])
        if el_count['1'] >= el_count['0']:
            inp = [x for x in inp if x[i]=='0']
        else:
            inp = [x for x in inp if x[i]=='1']
        i += 1
    return inp[0]

bin_to_dec(oxygen(inp))*bin_to_dec(co2(inp))
```

## Improved versions

```python
def part1(inp):
    bits = [[x[i] for x in inp] for i in range(len(inp[0]))]
    max_bits, min_bits = '', ''
    for i in range(len(bits)):
        max_bits += '1' if sum([x == '1' for x in bits[i]]) > len(inp)/2 else '0'
        min_bits += '0' if max_bits[-1] == '1' else '1'
    return int(max_bits, base=2)*int(min_bits, base=2)

part1(inp)
```

```python
def part2(inp):
    oxygen, co2 = inp, inp
    i=0
    while len(oxygen+co2) > 2:
        if len(oxygen)>1:
            oxygen = [x for x in oxygen if x[i] == '1'] if sum([x[i] == '1' for x in oxygen]) >= len(oxygen)/2 \
            else [x for x in oxygen if x[i] == '0']
        if len(co2)>1:
            co2 = [x for x in co2 if x[i] == '0'] if sum([x[i] == '0' for x in co2]) <= len(co2)/2 \
            else [x for x in co2 if x[i] == '1']
        i += 1
    return int(oxygen[0], base=2)*int(co2[0], base=2)

part2(inp)
```
